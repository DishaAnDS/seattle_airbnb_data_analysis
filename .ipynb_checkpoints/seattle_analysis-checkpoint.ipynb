### 3. Shared column - listing_id & id check

It seems the `listing_id` in calendar, `id` in listings, `listing_id` in reviews point to same thing. Let's check each column unique values and see whether they match each other or not.

```python
def unique_value_count(data):
    return len(data.unique())
```

```python
unique_value_count(calendar['listing_id'])
```

```python
unique_value_count(listings['id'])
```

```python
unique_value_count(reviews['listing_id'])
```

Since calendar `listing_id` has same unique value as `id` in listings, let's see whether they have identical values.

```python
set(calendar['listing_id']) == set(listings['id'])
```

So, calendar `listing_id` and listings `id` actually have identical values. Since reviews `listing_id` is less than the listings `id`. Review `listing_id` should be a part of entire id.

```python
set(reviews['listing_id']).issubset(listings['id'])
```

So reviews `listing_id` is a subset of the whole `id` for listings.

### 4. Missing Values Ratio Above 60%

```python
def missing_value_cols_ratio(df, ratio = 0.6):
    temp = pd.isnull(df).sum()/df.shape[0]>ratio
    missing_cols = set(temp[temp].index.values)
    return missing_cols
```

```python
missing_value_cols_ratio(calendar)
```

```python
missing_value_cols_ratio(listings,0.6)
```

The `license`, `monthly_price` and `square_feet` have more or equalt to 60% missing values.

### 3. Check is there seasonality of different prices when they are available.

```python

random.seed(1234)
```

```python
available_price = avai_t.copy()
# Remove the dollar sign $ and , for price
available_price['price']= available_price['price'].str.replace(',', '')
available_price['price']= available_price['price'].str.replace('$', '')

# Change the str to numerical value for price
available_price['price'] = available_price['price'].astype(float)
available_price_by_id = available_price.groupby(['listing_id']).mean()
```

```python
# Calculate each listing_id price average and then get the difference between price and mean of the price
avg_price = available_price_by_id.reset_index()
avg_price.columns = ['id', 'avg_price']
price_df = pd.merge(avg_price, available_price, how='left',  left_on='id', right_on = "listing_id")
price_df['price_diff'] = price_df['price'] - price_df['avg_price']
```

```python
# Randomly select 5 listing id, and check their results.
selected_id = list(avg_price.sample(n = 5)['id'])
```

```python
selected_df = price_df[price_df['id'].isin(selected_id)]
selected_df.loc[:, 'id'] = 'id_' + selected_df['id'].astype(str)
```

```python
# Random Picked 5 Listed Property Price Difference from Average Price
plt.figure(figsize=(25,10))
ax = sns.scatterplot(x="date", y="price_diff",
                     hue="id", 
                     data=selected_df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40)
plt.tight_layout()
plt.show()
```

From the above price, we can see that for some listing_id, price are very stable, but some are showing seasonality. This leads to my next exploration -- is there any features in listings for each id responsible for this kind of price behaviour?

## Modeling

Use the explored and related features to model on the price of the property and see how good the result we can get.

Let's take a closer look at the selected features: <br/>
`accommodates`, `bedrooms`, `beds`, `weekly_price`, `security_deposit`, `cleaning_fee`, `guests_included`, `host_listings_count`, `host_total_listings_count`, `room_type`

```python
room_type_df = pd.get_dummies(listings["room_type"],prefix='room_type')
```

```python
select_df = num_listings[['accommodates', 'bedrooms', 'beds', 
              'weekly_price', 'security_deposit', 'cleaning_fee',
              'guests_included', 'host_listings_count', 'host_total_listings_count', 
                          'calculated_host_listings_count', 'price']].copy()
```

```python
df = pd.concat([room_type_df, select_df], axis=1, join='inner')
df = df.dropna()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(df.corr(),annot=True)
```

Consider the highly correlated features, just keep one of them. So let's remove `room_type_Entire home/apt`, `bedrooms` and `beds`. `weekly_price` can be calculated by using `price`, so we need to remove it too.

```python
final_df = df.drop(['room_type_Entire home/apt', 'beds', 'bedrooms', 'weekly_price'], axis =1)
```

```python
final_df.columns
```

```python
#Split into explanatory and response variables
X = final_df.drop(['price'], axis = 1)
y = final_df['price']

#Split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .30, random_state=123) 
```

#### Linear Regression

```python
lm_model = LinearRegression(normalize=True) # Instantiate
# Fit the model
lm_model.fit(X_train, y_train) 
        
#Predict and score the model
y_test_preds = lm_model.predict(X_test) 
```

```python
print("R^2 is: {}".format(str(r2_score(y_test, y_test_preds))))
print("Mean Squared Error is: {}".format(str(mean_squared_error(y_test, y_test_preds))))
```

```python
# Let's visualize the distribution of bias we got for linear regression
lr_pred_df = pd.DataFrame({"lr_pred": y_test_preds, "lr_actual": y_test})
lr_pred_df['bias'] = lr_pred_df['lr_pred'] - lr_pred_df['lr_actual']
```

```python
lr_pred_df['bias'].hist(figsize=[10,5])
```

We can tell that most of the values are around 0 but with a relative big spread (around  -100 to 100) and there are some very extremely different prediction which needs further investigation.

#### Random Forest Regression

```python
regr = RandomForestRegressor()
# Fit the model
regr.fit(X_train, y_train)

#Predict and score the model
y_test_preds = regr.predict(X_test) 
```

```python
print("R^2 is: {}".format(str(r2_score(y_test, y_test_preds))))
print("Mean Squared Error is: {}".format(str(mean_squared_error(y_test, y_test_preds))))
```

```python
rf_pred_df = pd.DataFrame({"rf_pred": y_test_preds, "rf_actual": y_test})
rf_pred_df['bias'] =rf_pred_df['rf_pred'] - rf_pred_df['rf_actual']
```

```python
rf_pred_df['bias'].hist(figsize=[10,5])
```

Random forest model has a similar distribution.
